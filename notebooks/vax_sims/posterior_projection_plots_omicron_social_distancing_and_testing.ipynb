```python
import dill
from fall_2021_projection_utils import aggregate_trajs_student_only
```

```python

from os import listdir
from os.path import isfile, join

path = 'posterior_test_frequency_sims_1639006030/'
files = [join(path, f) for f in listdir(path) if isfile(join(path, f)) and 'omicron' in f]
```

```python
sampled_points = {}
aggregated_trajs = {}
count = 0
for f in files:
    if f in sampled_points:
        continue
    with open(f, "rb") as fhandle:
        [poind_idxs, point, inf_trajs_by_group] = dill.load(fhandle)
        sampled_points[f] = point
        aggregated_trajs[f] = aggregate_trajs_student_only(inf_trajs_by_group)
    count += 1
    if count % 50 == 0:
        print("Loaded {} points".format(count))
        #break
```

```python
def get_info(fname):
    base_fname =  fname.split('.dill')[0]
    remainder_fname, test_policy_idx = base_fname.split('_test_policy_idx_')
    more_remainder_fname, social_dist_multiplier = remainder_fname.split('_social_dist_multiplier_')
    omicron_multiplier = more_remainder_fname.split('_')[-1]
    return int(test_policy_idx), float(social_dist_multiplier), float(omicron_multiplier)

from test_policies import test_policy, policy_labels
import matplotlib.pyplot as plt
import numpy as np
fnames = sorted(list(aggregated_trajs.keys()))

xs_per_test_policy = {test_policy_idx:[] for test_policy_idx in test_policy}


# generate xs
for fname in fnames:
    test_policy_idx, social_dist_mult, omicron_mult = get_info(fname)
    xs_per_test_policy[test_policy_idx].append(social_dist_mult * omicron_mult)

for test_policy_idx in test_policy:
    xs_per_test_policy[test_policy_idx] = sorted(xs_per_test_policy[test_policy_idx])

# generate ys
mean_infs_per_test_policy = {test_policy_idx:[-1] * len(xs_per_test_policy[test_policy_idx]) 
                             for test_policy_idx in test_policy}
for fname in fnames:
    test_policy_idx, social_dist_mult, omicron_mult = get_info(fname)
    x_index = xs_per_test_policy[test_policy_idx].index(social_dist_mult * omicron_mult)
    mean_infs_per_test_policy[test_policy_idx][x_index] = np.mean([x.iloc[x.shape[0]-1] for x in aggregated_trajs[fname]])

plt.figure(figsize=(10,6))
for test_policy_idx in test_policy:
    plt.plot(xs_per_test_policy[test_policy_idx], mean_infs_per_test_policy[test_policy_idx],
            label=policy_labels[test_policy_idx], marker='o')
    
plt.legend(loc='best')
plt.xlabel("transmission multiplier")
plt.ylabel("average final infections")
plt.show()
    
```
