```python
import sys
import os
from io import StringIO
import pandas as pd
import numpy as np
module_path = os.path.abspath(os.path.join('.'))
sys.path.append(module_path + "/../../src/simulations_v2")
from stochastic_simulation import StochasticSimulation
from multi_group_simulation import MultiGroupSimulation
from load_params import load_params, load_multigroup_params
```

```python
group_params, group_names, interactions_mtx = load_multigroup_params("./in_person_multigroup.yaml")
# Adjust initial prevalence and test policy
group_params[0]['initial_ID_prevalence'] = 2 / group_params[0]['population_size']
group_params[1]['initial_ID_prevalence'] = 2 / group_params[1]['population_size']
group_params[0]['initial_ID_prevalence'] = 2 / group_params[0]['population_size']
group_params[1]['initial_ID_prevalence'] = 2 / group_params[1]['population_size']


multigroup_sim = MultiGroupSimulation(group_params, interactions_mtx)

sim_cum_dfs = list()
sim_individual_dfs = list()
nreps = 50
horizon = 22

for _ in range(nreps):
    sim_cum_dfs.append(multigroup_sim.run_new_trajectory(horizon)[1])
    individual_dfs = list()
    for idx, sim in enumerate(multigroup_sim.sims):
        individual_dfs.append(sim.sim_df)
```

```python
def get_cum_inf_trajectory(df):
    return np.sum(df[['cumulative_mild', 'cumulative_severe']], axis=1)

trajs = list()
for df in sim_cum_dfs:
    trajs.append(get_cum_inf_trajectory(df))
np.mean(trajs, axis=0)
```
